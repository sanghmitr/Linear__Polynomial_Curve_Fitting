```python
from numpy import arange
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot
```

```python
df = pd.read_excel('/content/drive/MyDrive/Curve_Fitting_RM/X-axis_Data.xlsx') 
x = []
x = df['X_axis'].values.tolist()
#print(x)

df = pd.read_excel('/content/drive/MyDrive/Curve_Fitting_RM/Y-axis_Data.xlsx') 
y = []
y = df['Y_axis'].values.tolist()
#print(y)
```

**Functions for curve equation generation**

```python
def linear(x, a, b):
	return a * x + b

def quadratic(x, a, b, c):
	return a * (x**2) + b * x + c

def cubic(x, a, b, c, d):
	return a * (x**3) + b * (x**2) + c * x + d
```

**Linear Curve Fit**

```python
# curve fit
popt, _ = curve_fit(linear, x, y)
# summarize the parameter values
a, b = popt
print('y = %.5f * x + %.5f' % (a, b))
# plot input vs output
pyplot.figure(figsize=(15,8))
pyplot.scatter(x, y)
# define a sequence of inputs between the smallest and largest known inputs
x_line = arange(min(x), max(x), 0.5)
# calculate the output for the range
y_line = linear(x_line, a, b)
# create a line plot for the mapping function
pyplot.xlabel("X-axis")
pyplot.ylabel("Y-axis")
pyplot.grid()
pyplot.plot(x_line, y_line, '--', color='red', linewidth=3)
pyplot.legend(['Linear curve'])
pyplot.show()
```

**Finding The Root Mean Square Error**

**Root Mean Square error and R2 score for Linear Curve**

```python
import math
from sklearn.metrics import mean_squared_error, r2_score
# Calculating Root Mean Squares Error
pred = []
rmse = 0
n = len(x)
for i in range(n):
    y_pred = b + (a * x[i])
    pred.append(y_pred)
    rmse += (y[i] - y_pred) ** 2
rmse = np.sqrt(rmse/n)
print("RMSE : ", rmse)
print("R2 score : %.2f" % r2_score(y,pred))
```

**Quadratic Curve Fit**

```python
popt, _ = curve_fit(quadratic, x, y)
# summarize the parameter values
a, b, c = popt
print('y = %.5f * x^2 + %.5f * x + %.5f' % (a, b, c))
# plot input vs output
pyplot.figure(figsize=(15,8))
pyplot.scatter(x, y)
# define a sequence of inputs between the smallest and largest known inputs
x_line = arange(min(x), max(x), 1)
# calculate the output for the range
y_line = quadratic(x_line, a, b, c)
# create a line plot for the mapping function
pyplot.xlabel("X-axis")
pyplot.ylabel("Y-axis")
pyplot.grid()
pyplot.plot(x_line, y_line, '--', color='red', linewidth=3)
pyplot.legend(['Quadratic curve'])
pyplot.show()
```

**Root Mean Square error and R2 score for Quadratic Curve**

```python
# Calculating Root Mean Squares Error
pred = []
rmse = 0
n = len(x)
for i in range(n):
    y_pred = c + b*x[i] + a * (x[i]**2)
    pred.append(y_pred)
    rmse += (y[i] - y_pred) ** 2
rmse = np.sqrt(rmse/n)
print("RMSE : ",rmse)
print("R2 score : %.2f" % r2_score(y,pred))
```

**Cubic Curve fit**

```python
popt, _ = curve_fit(cubic, x, y)
# summarize the parameter values
a, b, c, d = popt
print('y = %.5f * x^3 + %.5f * x^2 + %.5f*x + %.5f' % (a, b, c, d))
# plot input vs output
pyplot.figure(figsize=(15,8))
pyplot.scatter(x, y)
# define a sequence of inputs between the smallest and largest known inputs
x_line = arange(min(x), max(x), 1)
# calculate the output for the range
y_line = cubic(x_line, a, b, c, d)
# create a line plot for the mapping function
pyplot.xlabel("X-axis")
pyplot.ylabel("Y-axis")
pyplot.grid()
pyplot.plot(x_line, y_line, '--', color='red', linewidth=3)
pyplot.legend(['cubic curve'])
pyplot.show()
```

**Root Mean Square error and R2 score for Cubic Curve**

```python
# Calculating Root Mean Squares Error
pred = []
rmse = 0
n = len(x)
for i in range(n):
    y_pred = d + c*x[i] + b*(x[i]**2) + a * (x[i]**3)
    pred.append(y_pred)
    rmse += (y[i] - y_pred) ** 2
rmse = np.sqrt(rmse/n)
print("RMSE : ", rmse)
print("R2 score : %.2f" % r2_score(y,pred))
```
